## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Data Cleaning

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PTDL/WorldPopulation2023.csv")
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data['YearlyChange'] = data['YearlyChange'].str.replace("%",'')
data['YearlyChange'] = data['YearlyChange'].astype(float)
data['UrbanPop%'] = data['UrbanPop%'].replace('N.A.', '0%')
data['UrbanPop%'] = data['UrbanPop%'].str.replace("%",'')
data['UrbanPop%'] = data['UrbanPop%'].astype(float)
data['WorldShare'] = data['WorldShare'].str.replace("%",'')
data['WorldShare'] = data['WorldShare'].astype(float)
```

```python
data
```

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
```

## Univariate

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Bar plot for Region')
sns.countplot(x='Region', data=data, color='blue',
              order=data['Region'].value_counts().index)
plt.show()
```

## Bivariate

```python
grouped_data = data.copy()
grouped_data.loc[grouped_data['WorldShare'] < 2, 'Country'] = 'Other Countries'
grouped_data = grouped_data.groupby('Country')['WorldShare'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.pie(grouped_data['WorldShare'], labels=grouped_data['Country'], autopct='%1.1f%%', startangle=40)
plt.show()
```

```python
group_columns = ['Population2023', 'YearlyChange', 'NetChange', 'Density(P/Km²)', 'Land Area(Km²)',
                 'Migrants(net)', 'Fert.Rate', 'MedianAge', 'UrbanPop%', 'WorldShare']
fig, axarr = plt.subplots(5, 2, figsize=(15, 25))
for i, col in enumerate(group_columns):
    row = i // 2
    col = i % 2
    top5 = data.sort_values(by=group_columns[i], ascending=False).head(10)
    axarr[row][col].bar(top5['Country'], top5[group_columns[i]], color='lightgreen')
    axarr[row][col].set_title(f"Top 10 {group_columns[i]} By Country", fontsize=18)
    axarr[row][col].set_xlabel('Country', fontsize=14)
    axarr[row][col].set_ylabel(group_columns[i], fontsize=14)
    axarr[row][col].tick_params(axis='x', labelrotation=50)
plt.tight_layout()
plt.show()
```

```python
group_columns = ['Population2023', 'YearlyChange', 'NetChange', 'Density(P/Km²)',
                 'Land Area(Km²)', 'Migrants(net)', 'Fert.Rate', 'MedianAge']
fig, axarr = plt.subplots(4, 2, figsize=(15, 25))
for i, col in enumerate(group_columns):
    row = i // 2
    col = i % 2
    top5 = data.sort_values(by=group_columns[i]).head(10)
    axarr[row][col].bar(top5['Country'], top5[group_columns[i]], color='lightcoral')
    axarr[row][col].set_title(f"Top 10 Least {group_columns[i]} By Country", fontsize=18)
    axarr[row][col].set_xlabel('Country', fontsize=14)
    axarr[row][col].set_ylabel(group_columns[i], fontsize=14)
    axarr[row][col].tick_params(axis='x', labelrotation=50)
plt.tight_layout()
plt.show()
```

```python
population_by_region = data.groupby('Region')['Population2023'].sum().reset_index().sort_values(by='Population2023', ascending=False)
netchange_by_region = data.groupby('Region')['NetChange'].sum().reset_index().sort_values(by='NetChange', ascending=False)
migrants_by_region = data.groupby('Region')['Migrants(net)'].sum().reset_index().sort_values(by='Migrants(net)', ascending=False)

fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].bar(population_by_region['Region'], population_by_region['Population2023'])
axs[0].set_title('Population By Region')
axs[0].set_xlabel('Region')
axs[0].set_ylabel('Population')
axs[0].tick_params(axis='x', rotation=10)

axs[1].bar(netchange_by_region['Region'], netchange_by_region['NetChange'])
axs[1].set_title('NetChange By Region')
axs[1].set_xlabel('Region')
axs[1].set_ylabel('NetChange')
axs[1].tick_params(axis='x', rotation=10)

axs[2].bar(migrants_by_region['Region'], migrants_by_region['Migrants(net)'])
axs[2].set_title('Migrants(net) By Region')
axs[2].set_xlabel('Region')
axs[2].set_ylabel('Migrants(net)')
axs[2].tick_params(axis='x', rotation=10)
plt.tight_layout()
plt.show()
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(14, 12))
yc = data.groupby('Region')['YearlyChange'].mean().sort_values(ascending=False)
axarr[0, 0].bar(yc.index, yc.values)
axarr[0, 0].set_title(f"Mean YearlyChange by Region")
axarr[0, 0].set_xlabel('Region')
axarr[0, 0].set_ylabel('Mean Yearly Change')
axarr[0, 0].tick_params(axis='x', rotation=10)
fr = data.groupby('Region')['Fert.Rate'].mean().sort_values(ascending=False)
axarr[0, 1].bar(fr.index, fr.values)
axarr[0, 1].set_title(f"Mean Fert.Rate by Region")
axarr[0, 1].set_xlabel('Region')
axarr[0, 1].set_ylabel('Mean Fert.Rate')
axarr[0, 1].tick_params(axis='x', rotation=10)
ma = data.groupby('Region')['MedianAge'].mean().sort_values(ascending=False)
axarr[1, 0].bar(ma.index, ma.values)
axarr[1, 0].set_title(f"Mean MedianAge by Region")
axarr[1, 0].set_xlabel('Region')
axarr[1, 0].set_ylabel('Mean MedianAge')
axarr[1, 0].tick_params(axis='x', rotation=10)
up = data.groupby('Region')['UrbanPop%'].mean().sort_values(ascending=False)
axarr[1, 1].bar(up.index, up.values)
axarr[1, 1].set_title(f"Mean UrbanPop% by Region")
axarr[1, 1].set_xlabel('Region')
axarr[1, 1].set_ylabel('Mean UrbanPop%')
axarr[1, 1].tick_params(axis='x', rotation=10)
```

```python
dt = data.groupby('Region')['WorldShare'].sum().reset_index()
plt.figure(figsize=(5, 5))
plt.pie(dt['WorldShare'], labels=dt['Region'], autopct='%1.1f%%')
plt.title('WorldShare Distribution by Region')
plt.show()
```
